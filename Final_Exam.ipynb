```python
DATA STRUCTURE:
    tasks ← empty list

FUNCTION add_task():
    INPUT text ← GET LINE("Enter new task: ")
    text ← TRIM(text)
    IF text == "":
        DISPLAY "Task cannot be empty."
        RETURN
    APPEND text TO tasks
    DISPLAY "Added: " + text

FUNCTION list_tasks():
    IF LENGTH(tasks) == 0:
        DISPLAY "No tasks yet."
    ELSE:
        FOR i FROM 0 TO LENGTH(tasks)-1:
            DISPLAY STRING(i+1) + ". " + tasks[i]

FUNCTION remove_task():
    INPUT raw ← GET LINE("Enter task number to remove: ")
    TRY:
        index ← CONVERT raw TO INTEGER - 1
        DELETE tasks[index]
        DISPLAY "Task removed."
    EXCEPT:
        DISPLAY "Invalid number."

MAIN PROGRAM:
    LOOP:
        DISPLAY "1) Add  2) List  3) Remove  4) Quit"
        INPUT raw ← GET LINE("Choose 1–4: ")
        TRY:
            choice ← CONVERT raw TO INTEGER
        EXCEPT:
            DISPLAY "Please enter a number."
            CONTINUE LOOP
        IF choice == 1:
            CALL add_task()
        ELIF choice == 2:
            CALL list_tasks()
        ELIF choice == 3:
            CALL remove_task()
        ELIF choice == 4:
            DISPLAY "Goodbye!"
            BREAK LOOP
        ELSE:
            DISPLAY "Invalid choice."
```

Please help me that extend the Task Manager pseudocode by adding simple file persistence and defensive error handling while still remaining within the concepts from Weeks 1 to 10, using basic file operations (open, read, write) from Week 9, so that tasks would be loaded from a text file when the program started and saved back to the file after adding or removing a task. **AI’s Final Pseudocode (Excerpt):**

```python
DATA STRUCTURE:
    tasks ← empty list
    FILE_NAME ← "tasks.txt"

FUNCTION load_tasks():
    TRY:
        OPEN FILE_NAME FOR READING AS f
        FOR each line IN f:
            text ← TRIM(line)
            IF text != "":
                APPEND text TO tasks
        CLOSE f
    EXCEPT:
        DISPLAY "No existing file found."

FUNCTION save_tasks():
    TRY:
        OPEN FILE_NAME FOR WRITING AS f
        FOR each t IN tasks:
            WRITE t + "\n" TO f
        CLOSE f
    EXCEPT:
        DISPLAY "Could not save tasks."

FUNCTION add_task():
    INPUT text ← GET LINE("Enter new task: ")
    text ← TRIM(text)
    IF text == "":
        DISPLAY "Task cannot be empty."
        RETURN
    APPEND text TO tasks
    CALL save_tasks()
    DISPLAY "Added: " + text

FUNCTION list_tasks():
    IF LENGTH(tasks) == 0:
        DISPLAY "No tasks yet."
    ELSE:
        FOR i FROM 0 TO LENGTH(tasks)-1:
            DISPLAY STRING(i+1) + ". " + tasks[i]

FUNCTION remove_task():
    INPUT raw ← GET LINE("Enter task number to remove: ")
    TRY:
        index ← CONVERT raw TO INTEGER - 1
        DELETE tasks[index]
        CALL save_tasks()
        DISPLAY "Task removed."
    EXCEPT:
        DISPLAY "Invalid number."

MAIN PROGRAM:
    CALL load_tasks()
    LOOP:
        DISPLAY "1) Add  2) List  3) Remove  4) Quit"
        INPUT choice ← GET LINE("Choose 1–4: ")
        TRY:
            option ← CONVERT choice TO INTEGER
        EXCEPT:
            DISPLAY "Please enter a number."
            CONTINUE LOOP
        IF option == 1:
            CALL add_task()
        ELIF option == 2:
            CALL list_tasks()
        ELIF option == 3:
            CALL remove_task()
        ELIF option == 4:
            CALL save_tasks()
            DISPLAY "Goodbye!"
            BREAK LOOP
        ELSE:
            DISPLAY "Invalid choice."
```
