**My Initial Prompt:**

I need to design a simple command-line Task Manager that allows a user to add, list, and remove tasks stored in memory. Please provide pseudocode using only basic lists, functions, and while loops as taught in our course. Follow the six-step planning process.

**AI’s Pseudocode Response:**

```python
PROBLEM RESTATEMENT:
Design a CLI Task Manager that lets the user add, list, and remove tasks.

DATA STRUCTURE:
Use a list named tasks to store task strings.

FUNCTION add_task(task):
 APPEND task to tasks
 DISPLAY confirmation

FUNCTION list_tasks():
 IF tasks is empty → DISPLAY "No tasks yet"
 ELSE FOR each index, task in tasks → PRINT index + task

FUNCTION remove_task(index):
 TRY:
  DELETE tasks[index]
  DISPLAY "Task removed"
 EXCEPT:
  DISPLAY "Invalid index"

MAIN PROGRAM:
 WHILE True:
  DISPLAY menu options (1 Add / 2 List / 3 Remove / 4 Quit)
  GET choice from user
  IF choice == 1 → CALL add_task()
  ELIF choice == 2 → CALL list_tasks()
  ELIF choice == 3 → CALL remove_task()
  ELSE BREAK
```

I used the planning method from Week 2 - Computational Thinking and Algorithm Design, which teaches the six-step approach of defining the problem, analysing inputs and outputs, planning logic, implementing, testing, and reflecting.
