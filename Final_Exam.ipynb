## **1.2 Two Prompt Refinements (12 marks)**

**First Refinement Prompt:**

Could you please help me refine the pseudocode to include input validation so that invalid choices or empty tasks are handled gracefully? The pseudocode for my CLI Task Manager using only techniques taught in Weeks 1 to 10, such as lists for data storage, functions with parameters, loops for iteration, and basic error checking with try/except. The structured menu system requested that repeated until the user chose to quit, validated the menu choice so only numbers 1 to 4 were accepted, prevented adding empty tasks by checking the input string, and displayed task numbers starting from 1 for clarity. The goal was to apply the decision-making and loop patterns from Weeks 3 and 4, combined with the function and error-handling concepts from Week 6, to make the program more user-friendly and robust.

**AI’s Updated Pseudocode (Excerpt):**

```python
DATA STRUCTURE:
    tasks ← empty list

FUNCTION add_task():
    INPUT text ← GET LINE("Enter new task: ")
    text ← TRIM(text)
    IF text == "":
        DISPLAY "Task cannot be empty."
        RETURN
    APPEND text TO tasks
    DISPLAY "Added: " + text

FUNCTION list_tasks():
    IF LENGTH(tasks) == 0:
        DISPLAY "No tasks yet."
    ELSE:
        FOR i FROM 0 TO LENGTH(tasks)-1:
            DISPLAY STRING(i+1) + ". " + tasks[i]

FUNCTION remove_task():
    INPUT raw ← GET LINE("Enter task number to remove: ")
    TRY:
        index ← CONVERT raw TO INTEGER - 1
        DELETE tasks[index]
        DISPLAY "Task removed."
    EXCEPT:
        DISPLAY "Invalid number."

MAIN PROGRAM:
    LOOP:
        DISPLAY "1) Add  2) List  3) Remove  4) Quit"
        INPUT raw ← GET LINE("Choose 1–4: ")
        TRY:
            choice ← CONVERT raw TO INTEGER
        EXCEPT:
            DISPLAY "Please enter a number."
            CONTINUE LOOP
        IF choice == 1:
            CALL add_task()
        ELIF choice == 2:
            CALL list_tasks()
        ELIF choice == 3:
            CALL remove_task()
        ELIF choice == 4:
            DISPLAY "Goodbye!"
            BREAK LOOP
        ELSE:
            DISPLAY "Invalid choice."
```

**Course Reference:**

This refinement applied decision-making and menu loops from Weeks 3 and 4, functions and parameters from Week 4, and basic error handling with try and except from Week 6, to improve program usability within the scope of Weeks 1 and 10.

**Second Refinement Prompt:**

Please help me that extend the Task Manager pseudocode by adding simple file persistence and defensive error handling while still remaining within the concepts from Weeks 1 to 10, using basic file operations (open, read, write) from Week 9, so that tasks would be loaded from a text file when the program started and saved back to the file after adding or removing a task. It also requested the use of loops, conditions, and try and except to handle file errors gracefully without stopping the program. This iteration combined the structured thinking method from Week 7 with the file handling principles from Week 9 to make the program more reliable and practical.

**AI’s Final Pseudocode (Excerpt):**

```python
DATA STRUCTURE:
    tasks ← empty list
    FILE_NAME ← "tasks.txt"

FUNCTION load_tasks():
    TRY:
        OPEN FILE_NAME FOR READING AS f
        FOR each line IN f:
            text ← TRIM(line)
            IF text != "":
                APPEND text TO tasks
        CLOSE f
    EXCEPT:
        DISPLAY "No existing file found."

FUNCTION save_tasks():
    TRY:
        OPEN FILE_NAME FOR WRITING AS f
        FOR each t IN tasks:
            WRITE t + "\n" TO f
        CLOSE f
    EXCEPT:
        DISPLAY "Could not save tasks."

FUNCTION add_task():
    INPUT text ← GET LINE("Enter new task: ")
    text ← TRIM(text)
    IF text == "":
        DISPLAY "Task cannot be empty."
        RETURN
    APPEND text TO tasks
    CALL save_tasks()
    DISPLAY "Added: " + text

FUNCTION list_tasks():
    IF LENGTH(tasks) == 0:
        DISPLAY "No tasks yet."
    ELSE:
        FOR i FROM 0 TO LENGTH(tasks)-1:
            DISPLAY STRING(i+1) + ". " + tasks[i]

FUNCTION remove_task():
    INPUT raw ← GET LINE("Enter task number to remove: ")
    TRY:
        index ← CONVERT raw TO INTEGER - 1
        DELETE tasks[index]
        CALL save_tasks()
        DISPLAY "Task removed."
    EXCEPT:
        DISPLAY "Invalid number."

MAIN PROGRAM:
    CALL load_tasks()
    LOOP:
        DISPLAY "1) Add  2) List  3) Remove  4) Quit"
        INPUT choice ← GET LINE("Choose 1–4: ")
        TRY:
            option ← CONVERT choice TO INTEGER
        EXCEPT:
            DISPLAY "Please enter a number."
            CONTINUE LOOP
        IF option == 1:
            CALL add_task()
        ELIF option == 2:
            CALL list_tasks()
        ELIF option == 3:
            CALL remove_task()
        ELIF option == 4:
            CALL save_tasks()
            DISPLAY "Goodbye!"
            BREAK LOOP
        ELSE:
            DISPLAY "Invalid choice."
```

This refinement used file input and output from Week 9, loops and conditions from Weeks 3 and 4, and functions and error handling from Weeks 4 and 6 to add basic persistence and stability while staying within the Week 1–10 curriculum.

```python
# broken_task_manager_fixed.py
tasks = []

def add_task(task):
    """
    >>> tasks.clear(); add_task("Buy milk")
    Added: Buy milk
    """
    tasks.append(task)
    print(f"Added: {task}")

def remove_task(index):
    """
    >>> tasks.clear(); add_task("A"); add_task("B")
    Added: A
    Added: B
    >>> remove_task(1)
    Removed: B
    >>> remove_task(5)
    Invalid index!
    """
    # FIX 1 – correct 'len(tas)' → 'len(tasks)'
    # FIX 2 – add full bounds check
    if 0 <= index < len(tasks):
        removed = tasks[index]      # access after validation
        del tasks[index]
        print(f"Removed: {removed}")
    else:
        print("Invalid index!")

def list_tasks():
    """
    >>> tasks.clear(); list_tasks()
    No tasks available.
    >>> add_task("X"); list_tasks()
    Added: X
    1: X
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, t in enumerate(tasks):   # FIX 3 – use 'tasks' not 'task'
            print(f"{i+1}: {t}")

def main():
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")
    list_tasks()
    remove_task(1)
    list_tasks()

if __name__ == "__main__":
    main()
```
